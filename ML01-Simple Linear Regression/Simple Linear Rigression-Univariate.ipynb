#### Problem Statement
Given the home prices for few data points in 'homeprices.CSV', predict the home price for area=3300sq ft and area=5000 sq ft.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model

%matplotlib inline
```

```python
df=pd.read_csv('homeprices.csv')# Reading the data file
df.head()
```

```python
#Plotting Area vs price 
plt.scatter(df['area'],df['price'],color='red',marker='+')
plt.xlabel("Area in sq feets")
plt.ylabel("Prices in $")
plt.title("Area vs Price")
```

```python
reg=linear_model.LinearRegression()# Creating a object 'Reg' for a linear_model 
reg.fit(df[['area']],df['price'])# Training the regression model using the training data set, above data has 5 data points
```

Regression model is generated in above output, lets predict the prices for above 2 houses

```python
reg.predict([[3300]])
```

```python
# Lets see how our Linear regression line looks like
plt.xlabel("Area in (sq ft)")
plt.ylabel("Prices in USD($)")
plt.title("Linear Regression Line which fits data points")
plt.scatter(df['area'],df['price'],color='red',marker='+')
plt.plot(df['area'],reg.predict(df[['area']]))
```

Lets observe how it calculated exactly.
When you train your Linear Regression model it creats the values for the co-efficient(m) and intercept(c) in the Equation Y = mX + C.
lets see the values of m and C

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#Lets predict the price for house with area 500sqft
reg.predict([[5000]])
```

Till Now we predicted the prices by providing the area through manual input.
